```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
tax_df = pd.read_csv("grand_tax_dataset.csv")
happy_df = pd.read_csv("..\Happiness_Index_Grand.csv")
```

```python
happy_df
```

```python
tax_df
```

```python
merger = pd.merge(happy_df,tax_df, on = "Country Name",suffixes= {"_Happiness","_Tax_Revenue"})
merger
```

```python
plt.figure(figsize=(10,10))
ax = plt.scatter(x = merger[["2015_Tax_Revenue"]],
                   y = merger[["2015_Happiness"]])


x,y = merger['2015_Tax_Revenue'] , merger['2015_Happiness']
for i, txt in enumerate(merger['Country Name']):
    plt.annotate(txt,(x[i],y[i]))
    #print(i,txt,x[i],y[i],merger['2015_Tax_Revenue'][i],merger['2015_Happiness'][i])
```

```python
plt.figure(figsize=(10,10))
ax = plt.scatter(x = merger[["2016_Tax_Revenue"]],
                   y = merger[["2016_Happiness"]])


x,y = merger['2016_Tax_Revenue'] , merger['2016_Happiness']
for i, txt in enumerate(merger['Country Name']):
    plt.annotate(txt,(x[i],y[i]))
```

```python
# Creating the scatterplot

plt.figure(figsize=(10,10))

ax = plt.scatter(x = merger[["2017_Tax_Revenue"]],
                   y = merger[["2017_Happiness"]])

# Creating labels for the bubbles
x,y = merger['2017_Tax_Revenue'] , merger['2017_Happiness']
for i, txt in enumerate(merger['Country Name']):
    plt.annotate(txt,(x[i],y[i]))

# Labels     
plt.xlabel('Count of Rides',fontsize =17)
plt.ylabel('Average Fare ($) Per City', fontsize = 17)
plt.title('Average Fare ($) Per City, Count of rides, Driver Count (bubble size) and location (bubble colour)',fontsize = 16)
```
