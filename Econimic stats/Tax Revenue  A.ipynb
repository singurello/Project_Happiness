```python
import pandas as pd
import csv 
```

```python
tax_csv = pd.read_csv("tax_revenue.csv")
tax_chi_cbg = pd.read_csv("More_tax_rev_China_included.csv", skiprows =1)
```

```python
# TAX CHI 
tax_chi_cbg = tax_chi_cbg.reset_index()
```

```python
# TAX CHI 
tax_chi_cbg.columns = tax_chi_cbg.iloc[0]
tax_chi_cbg
```

```python
# TAX CHI 
tax_chi_cbg = tax_chi_cbg.drop([0])
tax_chi_cbg
```

```python
# Tax Chi
tax_chi_cbg = tax_chi_cbg.drop(tax_chi_cbg.columns[2:51], axis = 1)
tax_chi_cbg = tax_chi_cbg.set_index(["Country Name"], drop = True)
tax_chi_cbg = tax_chi_cbg.loc[['China','Cambodia','Brazil','Russian Federation']]
tax_chi_cbg.reset_index()
```

```python
tax_csv = tax_csv.drop(columns = {"Flag Codes","INDICATOR",'SUBJECT','FREQUENCY'})
```

```python
tax_csv = tax_csv.rename(columns = {"Value":"Tax Revenue (In Millions)"})
tax_csv.head()
```

```python
# Canada
CAN_CSV = tax_csv.loc[tax_csv["LOCATION"]=="CAN"].loc[tax_csv["MEASURE"]=="MLN_USD"]
CAN_CSV = CAN_CSV[42:52]
CAN_CSV
```

```python
# USA
USA_CSV = tax_csv.loc[tax_csv["LOCATION"]=="USA"].loc[tax_csv["MEASURE"]=="MLN_USD"]
USA_CSV = USA_CSV[42:52]
USA_CSV
```

```python
# Mexico
MEX_CSV = tax_csv.loc[tax_csv["LOCATION"]=="MEX"].loc[tax_csv["MEASURE"]=="MLN_USD"]
MEX_CSV = MEX_CSV[27:37]
MEX_CSV
```

```python
# Brazil
BRA_CSV = tax_chi_cbg.loc[["Brazil"]]
BRA_CSV
```

```python
# South Korea
KOR_CSV = tax_csv.loc[tax_csv["LOCATION"]=="KOR"].loc[tax_csv["MEASURE"]=="MLN_USD"]
KOR_CSV = KOR_CSV[35:45]
KOR_CSV
```

```python
# China 
CHI_CSV = tax_chi_cbg.loc[["China"]]
CHI_CSV
```

```python
# Cambodia
KHM_CSV = tax_chi_cbg.loc[["Cambodia"]]

KHM_CSV.head()
```

```python
DEU_CSV = tax_csv.loc[tax_csv["LOCATION"]=="DEU"].loc[tax_csv["MEASURE"]=="MLN_USD"]
DEU_CSV = DEU_CSV[28:38]
DEU_CSV
```

```python
RUS_CSV = tax_chi_cbg.loc[["Russian Federation"]]
RUS_CSV
```

```python
mergea = pd.concat([CAN_CSV,USA_CSV,KOR_CSV,MEX_CSV, DEU_CSV])
```
