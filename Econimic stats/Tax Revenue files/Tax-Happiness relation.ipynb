```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
tax_df = pd.read_csv("grand_tax_dataset.csv")
happy_df = pd.read_csv("..\Happiness_Index_Grand.csv")
```

```python
happy_df
```

```python
tax_df
```

```python
merger = pd.merge(happy_df,tax_df, on = "Country Name",suffixes= {"_Happiness","_Tax_Revenue"})
merger
```

```python
plt.figure(figsize=(10,10))
ax = plt.scatter(x = merger[["2015_Tax_Revenue"]],
                   y = merger[["2015_Happiness"]])


x,y = merger['2015_Tax_Revenue'] , merger['2015_Happiness']
for i, txt in enumerate(merger['Country Name']):
    plt.annotate(txt,(x[i],y[i]))
    #print(i,txt,x[i],y[i],merger['2015_Tax_Revenue'][i],merger['2015_Happiness'][i])

# Labels     
plt.xlabel('Tax Revenue (in Millions)',fontsize =16)
plt.ylabel('Happiness Index', fontsize = 16)
plt.title('Tax Revenue (in Millions) vs. Happiness Index: 2015',fontsize = 16)
plt.grid()

plt.show()
```
