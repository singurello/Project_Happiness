```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
unem_df = pd.read_csv("UNEM1.csv")
```

```python
unem_df
```

```python
plt.figure(figsize=(10,10))
ax = plt.scatter(x = unem_df[["2015_unem"]],
                   y = unem_df[["2015_happiness"]])


x,y = unem_df['2015_unem'] , unem_df['2015_happiness']
for i, txt in enumerate(unem_df['Country Name']):
    plt.annotate(txt,(x[i],y[i]))

# Labels     
plt.xlabel('Unemployment (in Precentage)',fontsize =16)
plt.ylabel('Happiness Index', fontsize = 16)
plt.title('Unemployment (in Precentage) vs. Happiness Index: 2015',fontsize = 16)
plt.grid()

plt.show()
```
