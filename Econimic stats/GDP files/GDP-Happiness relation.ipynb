```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
gdp_df = pd.read_csv('GDP1.csv')
```

```python
gdp_df
```

```python
plt.figure(figsize=(10,10))
ax = plt.scatter(x = gdp_df[["2015_gdp"]],
                   y = gdp_df[["2015_happiness"]])


x,y = gdp_df['2015_gdp'] , gdp_df['2015_happiness']
for i, txt in enumerate(gdp_df['Country Name']):
    plt.annotate(txt,(x[i],y[i]))
    
# Labels     
plt.xlabel('GDP (in Millions)',fontsize =16)
plt.ylabel('Happiness Index', fontsize = 16)
plt.title('GDP (in Millions) vs. Happiness Index: 2015',fontsize = 16)
plt.grid()

plt.show()
```
